```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.optimize import linear_sum_assignment

# Helper functions
def uniform_sampling(contour, num_points=100):
    """Uniformly sample points along the contour."""
    contour = contour.squeeze()
    length = cv2.arcLength(contour, True)
    sampled_points = []
    
    for i in np.linspace(0, length, num_points):
        point = cv2.pointPolygonTest(contour, (i, 0), True)
        if point >= 0:
            sampled_points.append([i, 0])
    
    return np.array(sampled_points)

def align_with_pca(points):
    """Align points using PCA."""
    pca = PCA(n_components=2)
    pca.fit(points)
    angle = np.arctan2(pca.components_[0][1], pca.components_[0][0])
    return angle

def create_cost_matrix(contours, reference_shapes):
    """Create cost matrix for Hungarian Matching."""
    num_contours = len(contours)
    num_references = len(reference_shapes)
    cost_matrix = np.zeros((num_contours, num_references))

    for i, contour in enumerate(contours):
        contour_points = uniform_sampling(contour)
        if contour_points.shape[0] == 0:
            cost_matrix[i, :] = np.inf  # Assign high cost if no points are sampled
            continue
        
        for j, ref_shape in enumerate(reference_shapes):
            ref_points = uniform_sampling(ref_shape)
            if ref_points.shape[0] == 0:
                cost_matrix[i, j] = np.inf  # Assign high cost if no points are sampled
            else:
                cost_matrix[i, j] = np.sum((contour_points - ref_points) ** 2)

    return cost_matrix

def hungarian_matching(cost_matrix):
    """Perform Hungarian Matching."""
    # Ensure no row or column is completely infeasible
    cost_matrix = np.where(np.isinf(cost_matrix), 1e9, cost_matrix)
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    return row_ind, col_ind

def determine_and_regularize_shape(contour, reference_shapes):
    """Determine and regularize the shape."""
    approx = cv2.approxPolyDP(contour, 0.02 * cv2.arcLength(contour, True), True)
    shape_name = determine_shape(approx)

    cost_matrix = create_cost_matrix([approx], reference_shapes)
    row_ind, col_ind = hungarian_matching(cost_matrix)

    # Ensure the cost is reasonable
    if cost_matrix[row_ind[0], col_ind[0]] < 1e5:
        return reference_shapes[col_ind[0]]
    else:
        return approx

def determine_shape(contour):
    """Determine shape name based on number of vertices."""
    num_vertices = len(contour)
    if num_vertices == 3:
        return 'Triangle'
    elif num_vertices == 4:
        return 'Rectangle'  # Assuming quadrilateral is a rectangle for simplicity
    elif num_vertices == 5:
        return 'Pentagon'
    elif num_vertices == 6:
        return 'Hexagon'
    elif num_vertices == 7:
        return 'Heptagon'
    elif num_vertices == 8:
        return 'Octagon'
    else:
        return 'Unknown'

def process_image(image_path, reference_shapes):
    """Process an image to identify and regularize shapes."""
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blurred, 50, 150)
    contours, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        regularized_shape = determine_and_regularize_shape(contour, reference_shapes)
        cv2.drawContours(image, [regularized_shape], -1, (0, 255, 0), 2)

    output_path = f"regularized_{image_path.split('/')[-1]}"
    cv2.imwrite(output_path, image)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Processed Image: {output_path}")
    plt.axis('off')
    plt.show()
    print(f"Processed and saved: {output_path}")

if __name__ == "__main__":
    # Define reference shapes
    reference_shapes = [
        np.array([[0, 0], [1, 0], [1, 1], [0, 1]], dtype=np.int32),  # Square
        np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]], dtype=np.int32)  # Triangle
    ]

    # Process images
    image_files = ["frag0.jpg", "frag1.jpg", "isolated.jpg", "occlusion2.jpg"]
    for image_file in image_files:
        process_image(image_file, reference_shapes)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.spatial import distance

def uniform_sampling(contour, num_points=100):
    """Uniformly sample points along the contour."""
    contour = contour.squeeze()
    length = cv2.arcLength(contour, True)
    num_samples = min(num_points, len(contour))
    sampled_points = []

    for i in np.linspace(0, length, num_samples):
        point = contour[int(i % len(contour))]
        sampled_points.append(point)
    
    return np.array(sampled_points)

def align_with_pca(points):
    """Align points using PCA."""
    pca = PCA(n_components=2)
    pca.fit(points)
    return pca

def create_reference_shapes():
    """Create reference shapes like square, triangle, etc."""
    return [
        np.array([[0, 0], [1, 0], [1, 1], [0, 1]], dtype=np.float32),  # Square
        np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]], dtype=np.float32),  # Triangle
        np.array([[0, 0], [2, 0], [2, 1], [0, 1]], dtype=np.float32),  # Rectangle
        np.array([[0, 0], [1, 0], [1.5, np.sqrt(3)/2], [0.5, np.sqrt(3)/2]], dtype=np.float32),  # Rhombus
        np.array([[0, 0], [1, 0], [1.5, np.sqrt(3)/2], [1, np.sqrt(3)], [0, np.sqrt(3)]], dtype=np.float32)  # Pentagon
    ]

def match_with_reference(contour_points, reference_shapes):
    """Match contour points with reference shapes."""
    min_distance = np.inf
    best_match = contour_points

    for ref_shape in reference_shapes:
        # Resample both shapes to have the same number of points
        num_points = min(len(contour_points), len(ref_shape))
        sampled_contour = contour_points[np.linspace(0, len(contour_points) - 1, num_points).astype(int)]
        sampled_ref_shape = ref_shape[np.linspace(0, len(ref_shape) - 1, num_points).astype(int)]

        # Calculate the distance between the shapes
        dist = np.sum((sampled_contour - sampled_ref_shape) ** 2)
        if dist < min_distance:
            min_distance = dist
            best_match = sampled_ref_shape

    return best_match

def align_and_regularize(contour, reference_shapes):
    """Align and regularize the contour."""
    contour_points = uniform_sampling(contour)
    pca = align_with_pca(contour_points)
    transformed_points = pca.transform(contour_points)

    # Match with reference shapes
    matched_shape = match_with_reference(transformed_points, reference_shapes)

    # Align back to the original coordinate system
    final_points = pca.inverse_transform(matched_shape)
    return final_points.astype(np.int32)

def process_image(image_path):
    """Process an image to identify and regularize shapes."""
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blurred, 50, 150)
    contours, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    reference_shapes = create_reference_shapes()

    for contour in contours:
        regularized_shape = align_and_regularize(contour, reference_shapes)
        cv2.drawContours(image, [regularized_shape], -1, (0, 255, 0), 2)

    output_path = f"regularized_{image_path.split('/')[-1]}"
    cv2.imwrite(output_path, image)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Processed Image: {output_path}")
    plt.axis('off')
    plt.show()
    print(f"Processed and saved: {output_path}")

if __name__ == "__main__":
    # List of images to process
    image_files = ["frag0.jpg", "frag1.jpg", "isolated.jpg", "occlusion2.jpg"]
    for image_file in image_files:
        process_image(image_file)
```
